#Question 3:
Your cousin is learning to code and wrote a program in Python. She clicked "Run"
and got an error saying "SyntaxError". She is confused and asks you — "I wrote everything correctly, why is the computer not understanding me?"

Explain to her why computers cannot directly understand what we write, and what happens between writing code and the computer actually running it.

#Answer:  
 We typically program by writing our code in specific formats so that humans will be able to comprehend it. Computers only recognize machine code, however, so there is another step in between us (the programmers) and the computer until we can accomplish tasks on our computer – using an interpreter to translate from Python to machine code.

When I wrote this program in Python and clicked Run, my Python interpreter processed each line of my written code (from top to bottom) in order to complete the entire task. As it interpreted the lines of code, it was looking out for any errors or mistakes such as spelling mistakes, incorrect symbols (missing symbols such as :, “”, (), etc.)

If the interpreter detected an error, it would report an error message that would say SyntaxError indicating that there had been an error in that line of code.

#Question 5:
Your younger brother opens his phone browser and types "www.flipkart.com". Within 2 seconds, the Flipkart homepage appears on his screen. He asks you — "How did this happen so fast? Where did this page come from?"

Explain the complete journey — from the moment he types the website name to the moment the page shows up on his screen. Use simple terms like DNS, server, browser, HTTP/HTTPS in your explanation.

#Answer:  
 When we type a website name into a browser, the Domain Name System (DNS) acts like a phonebook for the internet, locating the corresponding IP address of the site and providing it to the browser. The browser then communicates with the server at that IP address using an encrypted communication method. Today, most companies use HTTPS (Hypertext Transfer Protocol Secure) to ensure a more secure connection. Finally, the server sends the website content back to the browser for us to view.

Explain to him what cloud computing is, why it is useful for startups, and give a simple example of how cloud services work.

#Answer:  
Cloud services provide a way to store data on the internet. This approach is a cost-effective method for data storage because it eliminates the need for physical servers; instead, you can simply rent cloud services. For example, if our app experiences 1,000 client visits and we have our own server, we may need to purchase additional servers during peak times. However, once the traffic decreases, those extra servers will remain unused. In contrast, with cloud services, we can easily rent additional servers during high traffic periods and then relinquish them when the number of users returns to normal.

#Question 8:
What is the Software Development Life Cycle (SDLC)? Name all its phases and write one line about what happens in each phase.

#Answer:  
 The Software Development Life Cycle (SDLC) is a step-by-step process used to design, develop, test, and maintain software in a structured way.
 Phases of SDLC
1. Planning: Decide what the software should do, its goals, cost, and time required.
2. Requirement Analysis: Collect and understand what users need from the software.
3. Design: Create the blueprint of the software (how it will look and work).
4. Development (Coding): Programmers write the actual code to build the software.
5. Testing: Check the software for errors (bugs) and fix them.
6.  Deployment: Release the software so users can start using it.
7. Maintenance: Update and improve the software after release, fixing new issues if they appear.


#Question 9:
What is the difference between frontend development and backend development? Give one real example to show how both work together in a single application.

#Answer:  
The aspect of an application that users see – such as buttons, pages, images and layout – and can interface with directly to navigate through the application is called the frontend. The primary goal of the frontend is to provide a good user experience.

The aspect of an application that users do not see, but rather runs in the background, processes data, stores information, and handles the business rules behind the scenes, is called the backend. The backend makes sure that all aspects of the application function correctly and smoothly together.

For example: In an e-commerce (online shopping) application, the frontend contains the products (and their associated images) and the button to "Buy Now." When the user clicks the "Buy Now" button on the frontend, the backend will verify the availability, pricing and user information associated with the order. The backend then places the order and sends a confirmation of the order back to the frontend.

#Answer:

#HTTP (HyperText Transfer Protocol)
1. Sends data between your browser and website in plain text
2. Anyone who intercepts it can read the information
3. Not secure for sensitive data

#HTTPS (HyperText Transfer Protocol Secure)
1. Sends data in encrypted form (locked code)
2. Only the correct receiver can decode it
3. Much more secure and trustworthy

Why HTTPS is important for banking/payment websites is when we use banking or payment websites, we enter sensitive information like Bank details, Card numbers, OTPs and passwords
HTTPS is important because It protects your data from hackers, Prevents theft of passwords and money details, Stops attackers from changing data during transfer
